```python
import os
import pandas as pd

# Define the root directory where the 'Decomposed_data' folder is located
root_dir = 'Skeleton_analysis_full/Decomposed_data'

# Initialize a list to collect data
data = []

# Iterate through each treatment folder
for treatment in os.listdir(root_dir):
    treatment_dir = os.path.join(root_dir, treatment)
    if os.path.isdir(treatment_dir):  # Ensure it's a directory
        # Iterate through each subject folder in the current treatment folder
        for subject_folder in os.listdir(treatment_dir):
            subject_dir = os.path.join(treatment_dir, subject_folder)
            output_dir = os.path.join(subject_dir, 'output')
            if os.path.isdir(output_dir):  # Ensure the output folder exists
                # Process final_results.csv for junction counts
                final_results_path = os.path.join(output_dir, 'final_results.csv')
                if os.path.exists(final_results_path):
                    final_results_df = pd.read_csv(final_results_path)
                
                # Iterate through CSV files for each layer in the output folder
                for file in os.listdir(output_dir):
                    if file.endswith('.csv') and not file.startswith('final'):
                        # Extract layer information
                        layer_number = file.split('_layer_')[-1].split('.')[0]
                        
                        # Read the CSV file
                        csv_path = os.path.join(output_dir, file)
                        df = pd.read_csv(csv_path)
                        
                        # Calculate total branch length
                        total_length = df['Branch length'].sum()
                        
                        # Calculate average branch length
                        num_branches = len(df)
                        average_length = total_length / num_branches if num_branches > 0 else 0
                        
                        # Extract eye and subject ID from the subject folder name
                        eye = subject_folder.split('_')[1]
                        subject_id = subject_folder.split('_')[0]
                        
                        # Extract number of junctions for the corresponding layer from final_results_df
                        junctions_row = final_results_df[final_results_df['Filename'].str.contains(f'layer_{layer_number}.tif')]
                        if not junctions_row.empty:
                            number_of_junctions = junctions_row['Junctions'].values[0]
                        else:
                            number_of_junctions = None
                        
                        # Append data to the list
                        data.append([treatment, eye, subject_id, layer_number, total_length, number_of_junctions, average_length])

df_final = pd.DataFrame(data, columns=['Treatment', 'Eye', 'SubjectID', 'LayerNumber', 'Total Length', 'Number of Junctions', 'Average Branch Length'])
```

```python
# For Total Length
df_filtered_length = df.dropna(subset=['Total Length'])
df_pivot_length = df_filtered_length.pivot_table(index=['SubjectID', 'Treatment', 'LayerNumber'], columns='Eye', values='Total Length').reset_index()
df_pivot_length['TotalLengthDiff'] = df_pivot_length['OD'] - df_pivot_length['OE']
df_diff_length = df_pivot_length.melt(id_vars=['SubjectID', 'Treatment', 'LayerNumber'], value_vars=['TotalLengthDiff'], 
                                      var_name='Metric', value_name='Total Length').dropna()

df_diff_length = df_diff_length[['SubjectID', 'Treatment', 'LayerNumber', 'Total Length']]
df_diff_length['DeltaTotalLength'] = df_diff_length['Total Length']

# For Number of Junctions
df_filtered_junctions = df.dropna(subset=['Number of Junctions'])
df_pivot_junctions = df_filtered_junctions.pivot_table(index=['SubjectID', 'Treatment', 'LayerNumber'], columns='Eye', values='Number of Junctions').reset_index()
df_pivot_junctions['NumberOfJunctionsDiff'] = df_pivot_junctions['OD'] - df_pivot_junctions['OE']
df_diff_junctions = df_pivot_junctions.melt(id_vars=['SubjectID', 'Treatment', 'LayerNumber'], value_vars=['NumberOfJunctionsDiff'], 
                                            var_name='Metric', value_name='Number of Junctions').dropna()

df_diff_junctions = df_diff_junctions[['SubjectID', 'Treatment', 'LayerNumber', 'Number of Junctions']]
df_diff_junctions['DeltaNumberOfJunctions'] = df_diff_junctions['Number of Junctions']
```

```python
layers_of_interest = [-2, -1, 0, 1, 2]
df_layers = df[df['LayerNumber'].isin(layers_of_interest)]
metrics = ['Average Branch Length', 'Branchpoint Density', 'Vessel Length Density']
units = ['(# pixels)', '(#/pixel)', '(#/pixel)']
df_mean_metrics = df_layers.groupby(['SubjectID', 'Eye', 'Treatment']).agg({metric:'mean' for metric in metrics}).reset_index()

# Convert 'Treatment' to categorical type
df_mean_metrics['Treatment'] = pd.Categorical(df_mean_metrics['Treatment'])

sns.set(style="whitegrid", context="talk")

for metric in metrics:
    fig, ax = plt.subplots(figsize=(14, 8))
    box_plot = sns.boxplot(x="Treatment", y=metric, hue="Eye", data=df_mean_metrics, palette="colorblind", showfliers=False, ax=ax)
    strip_plot = sns.stripplot(x="Treatment", y=metric, hue="Eye", data=df_mean_metrics, dodge=True, jitter=0.1, marker='o', color='black', size=5, zorder=1, ax=ax)


    df_median_area = df_layers.groupby(['Treatment', 'Eye'])[metric].median().reset_index()

    print(f'Median {metric} values by Treatment and Eye')
    print(df_median_area)

    # Annotations
    texts = []
    for line in range(0, df_mean_metrics.shape[0]):
        point = ax.text(
            x=df_mean_metrics['Treatment'].cat.codes[line] + (-0.25 if df_mean_metrics['Eye'][line] == 'OD' else 0.25),  # Position offset
            y=df_mean_metrics[metric][line],
            s=df_mean_metrics['SubjectID'][line],  # The label to add
            horizontalalignment='center',
            size='medium',
            color='black',
            weight='semibold',
        )
        texts.append(point)

    adjust_text(texts, ax=ax)

    # Handle the legend and plot layout
    handles, labels = strip_plot.get_legend_handles_labels()
    ax.legend(handles[:2], labels[:2], title='Eye', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.set_xlabel('Treatment', fontsize=20)
    ax.set_ylabel(metric + f' {units[metrics.index(metric)]}', fontsize=20)

    plt.tight_layout()
    plt.show()
```

```python
metrics = ['Average Branch Length', 'Branchpoint Density', 'Vessel Length Density']

for metric in metrics:
    # Step 1: Filter for layers of interest
    layers_of_interest = [-2, -1, 0, 1, 2]
    df_layers = df[df['LayerNumber'].isin(layers_of_interest)]

    # Step 2: Calculate OD/OE ratio per layer for each SubjectID, Eye, and Treatment
    df_layer_pivot = df_layers.pivot_table(index=['SubjectID', 'Treatment', 'LayerNumber'], columns='Eye', values=metric).reset_index()
    df_layer_pivot.columns = ['SubjectID', 'Treatment', 'LayerNumber', f'{metric}_OD', f'{metric}_OE']

    df_layer_pivot[f'Ratio_OD_OE_{metric}'] = df_layer_pivot[f'{metric}_OD'] / df_layer_pivot[f'{metric}_OE']
    df_layer_pivot.replace([np.inf, -np.inf], np.nan, inplace=True)  # Cleaning up any infinite or NaN values from zero division

    # Step 3: Average these ratios by subject and treatment
    df_mean_ratios = df_layer_pivot.groupby(['SubjectID', 'Treatment']).agg({f'Ratio_OD_OE_{metric}':'mean'}).reset_index()

    # Calculate median ratios by treatment
    df_median_values = df_mean_ratios.groupby('Treatment')[f'Ratio_OD_OE_{metric}'].median().reset_index()

    print(f"Median 'Ratio OD/OE {metric}' values by 'Treatment':")
    print(df_median_values)

    # Step 4: Plotting
    plt.figure(figsize=(14, 8))
    sns.set(style="whitegrid", context="talk")

    box_plot = sns.boxplot(x="Treatment", y=f'Ratio_OD_OE_{metric}', data=df_mean_ratios, palette="colorblind", showfliers=False, width=0.5)
    strip_plot = sns.stripplot(x="Treatment", y=f'Ratio_OD_OE_{metric}', data=df_mean_ratios, jitter=0.05, marker='o', color='black', size=5, zorder=1)

    max_y = 4
    plt.ylim(0, max_y + max_y * 0.1)

    plt.xlabel('Treatment', fontsize=20)
    plt.ylabel(f'Ratio OD/OE {metric}', fontsize=20)

    # Adding text annotations for each point
    texts = []
    for index, row in df_mean_ratios.iterrows():
        x_position = list(df_mean_ratios['Treatment'].unique()).index(row['Treatment'])  # Adjust x position based on treatment
        texts.append(plt.text(x=x_position, y=row[f'Ratio_OD_OE_{metric}'], s=row['SubjectID'], color='black', ha='center', va='bottom'))

    # Automatically adjust text to minimize overlap
    adjust_text(texts)

    plt.tight_layout()
    plt.show()
```

# AUC diff OD-OE

```python
import pandas as pd
import numpy as np
from scipy.integrate import trapz


treatment_mapping = df[['SubjectID', 'Treatment']].drop_duplicates().set_index('SubjectID')


df_OD = df[df['Eye'] == 'OD']
df_OE = df[df['Eye'] == 'OE']

common_layers = df.groupby(['SubjectID', 'Eye'])['LayerNumber'].max().unstack().min(axis=1)

aucs = {}
for subject_id, min_layers in common_layers.items():
    for eye_type, df_eye in {'OD': df_OD, 'OE': df_OE}.items():
        subject_data = df_eye[(df_eye['SubjectID'] == subject_id) &
                              (df_eye['LayerNumber'] <= min_layers)]
        subject_data = subject_data.sort_values('LayerNumber')
        auc = trapz(subject_data['Total Length'], subject_data['LayerNumber'])
        aucs[(subject_id, eye_type)] = auc

fold_changes = {}
for subject_id in common_layers.index:
    fold_change = aucs[(subject_id, 'OD')] / aucs[(subject_id, 'OE')]
    fold_changes[subject_id] = fold_change

df_fold_changes = pd.DataFrame(list(fold_changes.items()), columns=['SubjectID', 'FoldChangeAUC'])

df_fold_change_length = df_fold_changes.merge(treatment_mapping, left_on='SubjectID', right_index=True)


##############

treatment_mapping = df[['SubjectID', 'Treatment']].drop_duplicates().set_index('SubjectID')


df_OD = df[df['Eye'] == 'OD']
df_OE = df[df['Eye'] == 'OE']

common_layers = df.groupby(['SubjectID', 'Eye'])['LayerNumber'].max().unstack().min(axis=1)

aucs = {}
for subject_id, min_layers in common_layers.items():
    for eye_type, df_eye in {'OD': df_OD, 'OE': df_OE}.items():
        subject_data = df_eye[(df_eye['SubjectID'] == subject_id) &
                              (df_eye['LayerNumber'] <= min_layers)]
        subject_data = subject_data.sort_values('LayerNumber')
        auc = trapz(subject_data['Number of Junctions'], subject_data['LayerNumber'])
        aucs[(subject_id, eye_type)] = auc

fold_changes = {}
for subject_id in common_layers.index:
    fold_change = aucs[(subject_id, 'OD')] / aucs[(subject_id, 'OE')]
    fold_changes[subject_id] = fold_change

df_fold_changes = pd.DataFrame(list(fold_changes.items()), columns=['SubjectID', 'FoldChangeAUC'])

df_fold_change_junctions = df_fold_changes.merge(treatment_mapping, left_on='SubjectID', right_index=True)

```
